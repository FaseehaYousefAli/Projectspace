```python
df.groupby("coffee_name")["money"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Total sales by coffee")
plt.xlabel("Coffee name")
plt.ylabel("Money(Rs)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.groupby("cash_type")["money"].sum().plot(kind="bar",color=["gold","skyblue"])
plt.title("Total Sales by Payment Type")
plt.xlabel("Payment Method")
plt.ylabel("Total Money")
plt.tight_layout()
plt.show()
```

```python
df["date"] = pd.to_datetime(df["date"])
df["month"] = df["date"].dt.to_period("M")
monthly_counts = df.groupby(["month", "coffee_name"]).size().reset_index(name="cups_sold")

print(monthly_counts)
```

```python
monthly_counts["month"] = monthly_counts["month"].astype(str)
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_counts, x="month", y="cups_sold", hue="coffee_name", marker="o")
plt.title("Monthly Sales Trend of Each Coffee")
plt.xlabel("Month")
plt.ylabel("Cups Sold")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title="Coffee Name", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df["weekday"] = df["date"].dt.day_name()
```

```python
weekday_counts = df["weekday"].value_counts().reindex(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
weekday_counts.plot(kind="bar", color="yellow", edgecolor="black")
plt.title("Coffee Sales by Weekday")
plt.xlabel("Weekday")
plt.ylabel("Cups Sold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
daily_sales = df["date"].value_counts().sort_index()
```

```python
plt.figure(figsize=(14, 6))
daily_sales.plot(kind="line")
plt.title("Daily Coffee Sales")
plt.xlabel("Date")
plt.ylabel("Cups Sold")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df["datetime"] = pd.to_datetime(df["datetime"])
```

```python
df["hour"] = df["datetime"].dt.hour
```

```python
hourly_sales = df["hour"].value_counts().sort_index()
plt.figure(figsize=(12, 6))
hourly_sales.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Hourly Coffee Sales")
plt.xlabel("Hour of Day")
plt.ylabel("Cups Sold")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
filtered = df[df["coffee_name"].isin(["Americano with Milk", "Latte"])]
```

```python
payment_counts = filtered.groupby(["coffee_name", "cash_type"]).size().unstack(fill_value=0)
payment_counts.plot(kind="bar", figsize=(8, 5), color=["maroon", "lightgreen"])
plt.title("Payment Type for Top Selling Coffees")
plt.xlabel("Coffee Name")
plt.ylabel("Number of Orders")
plt.xticks(rotation=0)
plt.legend(title="Payment Method")
plt.tight_layout()
plt.show()
```
